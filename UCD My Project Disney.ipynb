```python
#creating the loop which check the released year, created new column and assigning status

for index, row in disney.iterrows():
    if row['release_year'] > 2000:
        disney.loc[index,'Movie Status'] = "Not too old"
        print(row['release_year'],disney.loc[index,'Movie Status'])
    if row['release_year'] < 2000:    
        disney.loc[index,'Movie Status'] = "Classic"
        print(row['release_year'], disney.loc[index,'Movie Status'])
    
```

```python
#sorting only MOVIE titles between 2012 and 2022 year 

last10 = disney[(disney['release_year'] >= 2012) & (disney['release_year'] <= 2022) & (disney['type'] == 'MOVIE')]

last10


```

```python
#grouping by the age certification

last10.groupby('age_certification').count()
```

```python
#sorting by last 5 years

last5 = disney[(disney['release_year'] >= 2017) & (disney['release_year'] <= 2022) & (disney['type'] == 'MOVIE') & (disney['age_certification'] == 'PG')]

last5
```

```python
#1990- 1995 

oldgoodyears = disney[(disney['release_year'] >= 1990) & (disney['release_year'] <= 1995) & (disney['type'] == 'MOVIE')]
last5.groupby('production_countries').count()
x = oldgoodyears ['production_countries']
y = oldgoodyears ['release_year']


plt.figure(figsize=(17,6))

plt.scatter(x, y, label = 'Ranking',s = 100, color = 'blue',
            alpha = 0.5, marker = '.', edgecolors='black')

plt.ylabel('release_year', fontsize = 10)
plt.xlabel('production_countries', fontsize = 12)
plt.title('Countries production', fontsize = 8)
plt.grid(axis='y')
plt.legend()
plt.show()
plt.clf()

```

```python
#type of disney production

plt.bar(disney['type'].unique(), disney['type'].value_counts())
plt.xlabel("Types of productions", fontsize=10)
plt.ylabel("Number of prodcutions", fontsize=10)
plt.title("Types of the Disney's productions", fontsize=15)
plt.show()
```
